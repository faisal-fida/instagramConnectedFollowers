```python
import networkx as nx
import matplotlib.pyplot as plt

graph = nx.DiGraph()
```

```python
instagram_users = {
    "Alice": ["Bob", "Eve"],
    "Bob": ["Alice", "Eve"],
    "Charlie": ["Bob"],
    "Dave": ["Eve"],
    "Eve": ["Alice","Bob"]
}
```

```python
# Add nodes for each user
graph.add_nodes_from(instagram_users.keys())

# Add edges for the connections (followers and followings)
for user, connections in instagram_users.items():
    for connection in connections:
        graph.add_edge(user, connection)
```

```python
layout = nx.arf_layout(graph)

nx.draw_networkx(graph, pos=layout, with_labels=True, node_size=800, node_color='lightblue', edge_color='black')
plt.title('Instagram Connections')
plt.axis('off')
plt.show()
```
